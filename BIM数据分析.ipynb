# 数据处理

* 给df1增加每天消耗卡路里的列consume

```python
food_calorie_mid=food.describe().calorie.loc['50%']
```

```python
def food_to_calorie(x):
    line=x.split('、')
    global sum
    sum=0;
    for i in line:
        if food[food.food==i].empty:
#             print("this is null")
              sum+=food_calorie_mid
        else:
#             print(food[food.food==i]['calorie'].iloc[0])
#             print(type(food[food.food==i]['calorie'].iloc[0]))
            sum+=food[food.food==i]['calorie'].iloc[0]
    return sum
```

```python
df1['consum_breakfast']=df1['breakfast'].apply(food_to_calorie)
df1.head()
```

```python
df1['consum_lunch']=df1['lunch'].apply(food_to_calorie)
df1.head()
```

```python
df1['consum_dinner']=df1['dinner'].apply(food_to_calorie)
df1.head()
```

```python
df1['consum_day']=df1['consum_breakfast']+df1['consum_lunch']+df1['consum_dinner']
df1.head()
```

```python
df1.columns=['date','weight(kg)','breakfast','lunch','dinner','sport','input_breakfast','input_lunch','input_dinner','input_day']
df1.head()
```

* 为df1增加每天消耗卡路里的列

```python
def sport_to_calorie(x):
    global sum
    line = x.split('(')
    line[1] = line[1][:-3]
    sport_type = line[0]
    time_str = line[1]
    time = int(time_str)
    
    consum_speed=sport[sport.sport=='跑步']['consume（kcal/m）'].iloc[0]
    sum=consum_speed*time
    return sum
```

```python
df1.head()
```

```python
df1['consum_day']=df1['sport'].apply(sport_to_calorie)
df1.head()
```

* 增加一列表示摄入热量减去运动消耗热量的差值

```python
df1['input_sub_consum_day']=df1['input_day']-df1['consum_day']
df1.head()
```

```python
df2=df1.loc[:,['date','weight(kg)','input_breakfast','input_lunch','input_dinner','input_day','consum_day','input_sub_consum_day']]
df2.head()
```

* 下图展现了早餐、午餐、晚餐分别占每天总摄入量的比例
> 三餐的摄入量比较平均，午餐为最高，约占37%，早餐和晚餐基本一致

```python
input_breakfast_sum=df2['input_breakfast'].sum()
input_lunch_sum=df2['input_lunch'].sum()
input_dinner_sum=df2['input_dinner'].sum()
input_day_sum=df2['input_day'].sum()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Pie

a=input_dinner_sum/input_day_sum
b=input_lunch_sum/input_day_sum
c=input_dinner_sum/input_day_sum


x_data = ["早餐", "午餐", "晚餐"]
y_data = [a,b,c]
data_pair = [list(z) for z in zip(x_data, y_data)]
data_pair.sort(key=lambda x: x[1])

(
    Pie(init_opts=opts.InitOpts(bg_color="#2c343c"))
    .add(
        series_name="访问来源",
        data_pair=data_pair,
        rosetype="radius",
        radius="55%",
        center=["50%", "50%"],
        label_opts=opts.LabelOpts(is_show=False, position="center"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="三餐摄入热量所占的比例",
            pos_left="center",
            pos_top="20",
            title_textstyle_opts=opts.TextStyleOpts(color="#fff"),
        ),
        legend_opts=opts.LegendOpts(is_show=False),
    )
    .set_series_opts(
        tooltip_opts=opts.TooltipOpts(
            trigger="item", formatter="{a} <br/>{b}: {c} ({d}%)"
        ),
        label_opts=opts.LabelOpts(color="rgba(255, 255, 255, 0.3)"),
    )
    .render_notebook()
)
```
